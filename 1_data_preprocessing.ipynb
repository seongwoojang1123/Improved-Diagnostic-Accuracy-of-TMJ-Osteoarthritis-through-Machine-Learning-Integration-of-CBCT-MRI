```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = '/content/drive/MyDrive/TMJ OA/20250709 CBCT MRI 데이터 추가분 통합.xlsx'

df = pd.read_excel(file_path, sheet_name='정리본')

# 'SEX (여성-F, 남성-M)' 컬럼명을 'SEX'로 변경
df.rename(columns={'SEX (여성-F, 남성-M)': 'SEX'}, inplace=True)
df['SEX'] = df['SEX'].map({'F': 0, 'M': 1})

# 빈문자열을 NaN으로 변경하기
df.replace("", np.nan, inplace=True)

columns = df.columns
print(len(columns))
print(columns)
df.head()
```

```python
df.describe()
```

```python
# CBCT와 MRI 이미지 갯수 확인

# CBCT 관련 컬럼들기
cbct_columns = ['CBCT_Rt OA (Osteoarthritis) (0=normal, 1=OA)', 'CBCT_Lt OA  (Osteoarthritis) (0=normal, 1=OA)']

# MRI 관련 컬럼들
mri_columns = ['MRI_Rt OA (0=normal, 1=OA)', 'MRI_Lt OA (0=normal, 1=OA)']

# CBCT 값 개수 확인 (0과 1의 개수)
cbct_counts = df[cbct_columns].apply(pd.Series.value_counts)
print("CBCT 데이터 개수:")
print(cbct_counts)

# MRI 값 개수 확인 (0과 1의 개수)
mri_counts = df[mri_columns].apply(pd.Series.value_counts)
print("\nMRI 데이터 개수:")
print(mri_counts)
```

```python
# 데이터 저장
data = pd.DataFrame(df)
print(f'data.shape : {data.shape}')
data.to_csv('/content/drive/MyDrive/TMJ OA/250825_TMJ OA data.csv', index=False, encoding = 'utf-8')
```
